```python
import pandas as pd
from datetime import datetime
import pandas_bokeh
```

```python
mydateparser = lambda x: datetime.strptime(x, "%d/%m/%Y" )
datos = pd.read_csv('innerjoins.csv', parse_dates=['fecha'],date_parser=mydateparser)
df = pd.DataFrame(datos)
df
```

```python
# Borro las columnas que no me sirven
df.drop(['code','name','pop_dens','country_percentage','dia_inicio','dia_cuarentena','ciudad','transmision_tipo','informe_tipo','fuente'], axis='columns', inplace=True)
df
```

```python
import numpy as np

from bokeh.io import curdoc, show
from bokeh.models import ColumnDataSource, Grid, Line, LinearAxis, Plot

N = 30
y = df['nue_casosconf']
x = df['fecha']

source = ColumnDataSource(dict(x=x, y=y))

plot = Plot(
    title=None, plot_width=300, plot_height=300,
    min_border=0, toolbar_location=None)

glyph = Line(x="x", y="y", line_color="#f46d43", line_width=6, line_alpha=0.6)
plot.add_glyph(source, glyph)

xaxis = LinearAxis()
plot.add_layout(xaxis, 'below')

yaxis = LinearAxis()
plot.add_layout(yaxis, 'left')

plot.add_layout(Grid(dimension=0, ticker=xaxis.ticker))
plot.add_layout(Grid(dimension=1, ticker=yaxis.ticker))

curdoc().add_root(plot)

show(plot)
```

```python
from bokeh.plotting import figure

# Create some data.
y = df['nue_casosconf']
x = df['fecha']
z = df['provincia']

# Create a new plot with a title and axis labels
p = figure(title="Simple Line Plot in Bokeh", x_axis_label='x', y_axis_label='y')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="Value", line_width=3)

# Show the results
show(p)
```

```python
from math import pi

p = figure(x_axis_type="datetime", title="CORONAvirus", plot_height=350, plot_width=800)
p.xaxis.formatter.days = '%m/%d/%Y'
p.xaxis.major_label_orientation = pi/3

p.line(df['fecha'], df['nue_casosconf'])

show(p)
```

```python
## grafico para leyenda interactiva 

from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import transform
x = df['fecha']
y = df['nue_casosconf']
color_mapper = LinearColorMapper(palette="Viridis256", low=df['nue_casosconf'].min(), high=df['nue_casosconf'].min())

p = figure(x_axis_label='Fecha', y_axis_label='NC', tools='', toolbar_location=None)
p.line(x, y, color=transform('weight', color_mapper), size=20, alpha=0.6, source=df)

color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12, location=(0,0), title='Weight')
p.add_layout(color_bar, 'right')

show(p)
```
