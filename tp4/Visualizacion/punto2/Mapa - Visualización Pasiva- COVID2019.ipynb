```python
import folium
import pandas as pd
```

```python
# Borro las columnas que no me sirven

datos = pd.read_csv('innerjoins.csv')
df = pd.DataFrame(datos)
df.drop(['code','name','pop_dens','country_percentage','dia_inicio','dia_cuarentena','ciudad','transmision_tipo','informe_tipo','fuente'], axis='columns', inplace=True)
df
```

```python
casos_cada_100_mil = []
casos_cada_100_mil = (df['nue_casosconf']/df['total_pop'])
df['casos_cada_100_mil'] = casos_cada_100_mil
df
```

```python
#Calcular la cantidad de casos por provincia
group_a = df.groupby('provincia')

tot_casos_p = []
for key, item in group_a:
    tot_casos_p.append({"provincia":key, "total_casos":item['nue_casosconf'].sum(), "penetracion":item['casos_cada_100_mil'].sum()})
```

```python
bsas_caba = tot_casos_p[0]['total_casos']+tot_casos_p[0]['total_casos']
bsas_caba_p = tot_casos_p[0]['penetracion']+tot_casos_p[0]['penetracion']
tot_casos_p.append({'provincia': 'Buenos Aires CABA', 'total_casos':bsas_caba, 'penetracion':bsas_caba_p})
tot_casos_p
```

```python
def definir_color(valor):
    color =""
    if valor > 0.001 :
        color = "#FF0000"
    if valor < 0.001 and valor > 0.0001:
        color = "#FF8000"
    if valor < 0.0001:
        color = "#FFFF00"
    if valor == 0.0:
        color = "#00FF00"
    
    return color
    
```
