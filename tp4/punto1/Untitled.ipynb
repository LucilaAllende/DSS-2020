```python
import tweepy
import json


def limpieza(tweet):
    mapeo = {
        ord('\t'): ' ',
        ord('\f'): ' ',
        ord('\r'): None
    }

    stripped_tweet = tweet.translate(mapeo)

    return stripped_tweet


# 4 cadenas para la autenticacion
consumer_key = ""
consumer_secret = ""
access_token = ""
access_token_secret = ""

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

# con este objeto realizaremos todas las llamadas al API
api = tweepy.API(auth,
                 wait_on_rate_limit=True,
                 wait_on_rate_limit_notify=True)


argentina = 23424747
# trends1 = api.trends_place(1)
# trends = set([trend['name'] for trend in trends1[0]['trends']])
# print(trends)

WORLD_WOE_ID = 1
# obtengo los primeros diez TT
world_trends = api.trends_place(argentina)
trends = []
tweets_limpios = []
# Selecionno solo 3 TT
for i in range(3):
    print(world_trends[0]['trends'][i][u'name'])
    trends.append(world_trends[0]['trends'][i][u'name'])

# Guardar datos en json
with open('data_limpia.json', 'w') as outfile:
    for trend in range(3):
        for tweet in tweepy.Cursor(api.search, q=trends[trend], tweet_mode="extended").items(2):
            tweets = json.dumps(tweet._json,  indent=4)
    # limpiar tweets
    for tweet in tweets:
        tweet_limpio = limpieza(tweet)
        tweets_limpios.append(tweet_limpio)

    json.dump(tweets_limpios, outfile, indent=4)

print(len(tweets_limpios))
print(tweets_limpios)
```
